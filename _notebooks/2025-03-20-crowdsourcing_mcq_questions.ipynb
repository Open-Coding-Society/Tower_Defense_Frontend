# Crowdsourcing Multiple-Choice Questions

---

### 1. What is one key benefit of crowdsourcing?

- A) It limits diversity in the contributions.  
- B) It can reduce computer bias by involving a wide range of perspectives.  
- C) It discourages community engagement.  
- D) It makes crowdsourcing efforts more expensive.  

<details>
  <summary>Click to reveal the correct answer</summary>
  **Correct Answer:** B) It can reduce computer bias by involving a wide range of perspectives.
</details>

---

### 2. Which of the following is an example of "crowd creation" in crowdsourcing?

- A) Kickstarter campaigns for funding new products.  
- B) Wikipedia, where users collaborate to create content.  
- C) Reddit upvotes for ranking content.  
- D) SETI@home using volunteer computing power.  

<details>
  <summary>Click to reveal the correct answer</summary>
  **Correct Answer:** B) Wikipedia, where users collaborate to create content.
</details>

---

### 3. What is a challenge commonly faced in data crowdsourcing?

- A) Ensuring the involvement of a few experts.  
- B) Managing intellectual property rights.  
- C) Ensuring data accuracy and reliability from diverse contributors.  
- D) Making the data collection process slow.  

<details>
  <summary>Click to reveal the correct answer</summary>
  **Correct Answer:** C) Ensuring data accuracy and reliability from diverse contributors.
</details>

---

### 4. Which of the following is a successful example of "crowd wisdom"?

- A) Kickstarter, where people contribute to funding a product.  
- B) Waze, where drivers share real-time traffic data.  
- C) Duolingo, where volunteers help translate language content.  
- D) Reddit, where users vote on the most relevant posts.  

<details>
  <summary>Click to reveal the correct answer</summary>
  **Correct Answer:** B) Waze, where drivers share real-time traffic data.
</details>

---

### 5. Which type of crowdsourcing is used to fund a new startup by collecting small contributions from a large group of people?

- A) Crowd voting  
- B) Crowd creation  
- C) Crowdfunding  
- D) Crowd wisdom  

<details>
  <summary>Click to reveal the correct answer</summary>
  **Correct Answer:** C) Crowdfunding
</details>
